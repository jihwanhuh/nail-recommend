```python
import pandas as pd

# íŒŒì¼ëª…
csv_path = "basic_data/ë¸”ë¡œê·¸ ë§í¬ ëª©ë¡.csv"

# ì¡°ì‚¬í•œ íƒ­ì„ ìœ„ íŒŒì¼ëª…ì˜ csv ë¡œ ë‹¤ìš´ë¡œë“œ
# ì²«ë²ˆì§¸ ì„¤ëª… í–‰ì€ ìˆ˜ë™ìœ¼ë¡œ ì‚­ì œí›„ ì‹¤í–‰

df_links = pd.read_csv(csv_path, encoding="cp949")
df_links = df_links.rename(columns={"Unnamed: 0": "ë²ˆí˜¸"})

df_links
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import mimetypes

BASE_FOLDER = "blog_images"   # ì´ë¯¸ì§€ ëª¨ì„ í´ë”
os.makedirs(BASE_FOLDER, exist_ok=True)

COMMON_HEADERS = {
    "User-Agent": "Mozilla/5.0"
}

def crawl_naver_blog(row_no: int, url: str):
    """
    row_no : CSV 'ë²ˆí˜¸' ì—´ ê°’ (1,2,3,...)
    url    : URL
    return : dict(title=..., body=..., image_count=...)
    """

    # 0) ê²‰ í˜ì´ì§€
    resp = requests.get(url, headers=COMMON_HEADERS)
    resp.raise_for_status()
    html = resp.text
    outer_url = resp.url

    soup = BeautifulSoup(html, "html.parser")

    # 1) iframe ì°¾ê¸°
    iframe = soup.select_one("iframe#mainFrame")
    if iframe is None:
        for f in soup.find_all("iframe"):
            src = f.get("src", "")
            if "PostView" in src:
                iframe = f
                break

    if iframe is None:
        print(f"[{row_no}] ë³¸ë¬¸ iframe ëª» ì°¾ìŒ")
        return {"title": "", "body": "", "image_count": 0}

    inner_src = iframe.get("src")
    inner_url = urljoin(outer_url, inner_src)

    # 2) iframe ì•ˆìª½ ìš”ì²­
    resp2 = requests.get(inner_url, headers=COMMON_HEADERS)
    resp2.raise_for_status()
    inner_html = resp2.text
    soup2 = BeautifulSoup(inner_html, "html.parser")

    # 3) ì œëª©
    title_tag = soup2.find("title")
    title = title_tag.get_text(strip=True) if title_tag else ""

    # 4) ë³¸ë¬¸ í…ìŠ¤íŠ¸
    paragraphs = [p.get_text(" ", strip=True) for p in soup2.find_all("p")]
    body_text = "\n".join(paragraphs)

    if len(body_text.strip()) == 0:
        main_div = soup2.select_one("div.se-main-container")
        if main_div:
            paragraphs = [
                p.get_text(" ", strip=True)
                for p in main_div.find_all(["p", "span"])
            ]
            body_text = "\n".join(paragraphs)

    # 5) ë³¸ë¬¸ ì´ë¯¸ì§€ (ë³¸ë¬¸ ì˜ì—­ + postfiles.pstatic.netë§Œ)
    content = (
        soup2.select_one("div.se-main-container")
        or soup2.select_one("#postViewArea")
        or soup2
    )
    img_tags = content.find_all("img")
    print(f"[{row_no}] ë³¸ë¬¸ ì˜ì—­ img íƒœê·¸ ê°œìˆ˜:", len(img_tags))

    image_count = 0
    for img_idx, img in enumerate(img_tags, start=1):
        src = img.get("data-origin") or img.get("data-src") or img.get("src")
        if not src:
            continue

        # ë„¤ì´ë²„ ë³¸ë¬¸ ì´ë¯¸ì§€ ì„œë²„ë§Œ (ì•„ì´ì½˜/ì´ëª¨í‹°ì½˜ ì œì™¸ìš©)
        if "postfiles.pstatic.net" not in src:
            continue

        if src.startswith("//"):
            src = "https:" + src
        elif src.startswith("/"):
            src = urljoin(inner_url, src)
        if src.startswith("data:"):
            continue

        # ì¸ë„¤ì¼ì´ë©´ í° íƒ€ì…ìœ¼ë¡œ ì‹œë„
        hi_src = src
        if "type=w80_blur" in src:
            hi_src = src.replace("type=w80_blur", "type=w966")

        img_headers = {
            "User-Agent": "Mozilla/5.0",
            "Referer": inner_url,
        }

        img_resp = None
        for candidate in [hi_src, src]:
            try:
                r = requests.get(candidate, headers=img_headers, timeout=10)
                r.raise_for_status()
                img_resp = r
                break
            except Exception:
                img_resp = None

        if img_resp is None:
            continue

        content_type = img_resp.headers.get("Content-Type", "")
        if "image" not in content_type:
            continue

        ext = mimetypes.guess_extension(content_type.split(";")[0]) or ".jpg"

        # íŒŒì¼ ì´ë¦„: img_ë²ˆí˜¸_ì´ë¯¸ì§€ë²ˆí˜¸.ext
        filename = os.path.join(
            BASE_FOLDER,
            f"img_{row_no:03d}_{img_idx:03d}{ext}"
        )

        with open(filename, "wb") as f:
            f.write(img_resp.content)

        image_count += 1
        print("ì €ì¥:", filename)

    return {
        "title": title,
        "body": body_text,
        "image_count": image_count,
    }
```

```python
results = []

for _, row in df_links.iterrows():
    row_no = int(row["ë²ˆí˜¸"])
    url = row["ë§í¬"]

    # ë§í¬ê°€ ë¹„ì–´ìˆëŠ” ê²½ìš° ì²˜ë¦¬
    if not isinstance(url, str) or url.strip() == "":
        print(f"[{row_no}] ë§í¬ ì—†ìŒ â†’ ê±´ë„ˆëœ€")
        results.append({
            "ë²ˆí˜¸": row_no,
            "ê²€ìƒ‰í•œ ê³³": row.get("ê²€ìƒ‰í•œ ê³³", ""),
            "ê²€ìƒ‰ì–´": row.get("ê²€ìƒ‰ì–´", ""),
            "ë§í¬": "",
            "ë¹„ê³ ": row.get("ë¹„ê³ ", ""),
            "title": "",
            "body": "",
            "image_count": 0
        })
        continue

    print(f"\n=== [{row_no}] í¬ë¡¤ë§ ì‹œì‘ ===")
    info = crawl_naver_blog(row_no, url)

    result_row = {
        "ë²ˆí˜¸": row_no,
        "ê²€ìƒ‰í•œ ê³³": row.get("ê²€ìƒ‰í•œ ê³³", ""),
        "ê²€ìƒ‰ì–´": row.get("ê²€ìƒ‰ì–´", ""),
        "ë§í¬": url,
        "ë¹„ê³ ": row.get("ë¹„ê³ ", ""),
        "title": info["title"],
        "body": info["body"],
        "image_count": info["image_count"],
    }
    results.append(result_row)

df_result = pd.DataFrame(results)
df_result
```
